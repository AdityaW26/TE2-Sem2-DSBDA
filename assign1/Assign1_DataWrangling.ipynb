```python
import pandas as pd
```

```python
df = pd.read_csv("melb_data.csv")
```

```python
print(df)
```

```python
sum1 = df['Rooms'].sum()       # calculating sum of rooms
count1 = df['Rooms'].count()       # calculating Count of different rooms
max1 = df['Rooms'].max()       # calculating max of rooms
min1 = df['Rooms'].min()       # calculating min of rooms
mean1 = df['Rooms'].mean()       # calculating mean of rooms
median1 = df['Rooms'].median()       # calculating median of rooms
print("Total rooms - " + str(sum1))       
print("Count of different rooms - " + str(count1))
print("Max rooms - " + str(max1))
print("Min rooms - " + str(min1))
print("Mean of rooms - " + str(mean1))
print("Median of rooms - " + str(median1))
```

```python
import matplotlib.pyplot as plt
```

```python
x = df.Postcode
y = df.Price
plt.plot(x,y)  # plotting a graph of Price vs Postcode
plt.show()
```

```python
x = df.YearBuilt
y = df.BuildingArea
plt.plot(x,y)  # plotting a graph of BuildingArea vs YearBuilt
plt.show()
```
